```python
import cobra, libsbml, collections, pyDOE, math
import pandas as pd
import numpy as np
from cobrafunctions.buildingediting import *
from cobrafunctions.analysing import *
from cobrafunctions.adjustedcobrafunctions import *
from tqdm import tqdm_notebook as tqdm

savelocation, inputlocation = nameexperiment("figure3") # Set name for files

model = cobra.io.read_sbml_model("Inputs/4_stage_GC.xml")

model = constrainmaintenance(model) #add maintenance
```

### Add import pseudometabolites

```python
importdict = {
    "MAL" : {
        "MAL_ce" : -1,
        "MAL_ATPASE_ec" : 1,
    },
    "SUCROSE" : {
        "SUCROSE_PROTON_ec" : 1,
        "SUCROSE_ce" : -1,
    },
    "GLC" : {
        "GLC_PROTON_ec" : 1, 
        "GLC_ce" : -1,       
    },
    "K" : {
        "K_ec" : 1,
        "K_ce" : -1,
        "K_PROTON_ec" : 1,
    },
    "FRU" : {
        "FRU_PROTON_ec" : 1,
        "FRU_ce" : -1,
    },
    "STARCH" : {},
    "Cl" : {
        "Cl_PROTON_ec" : 2,
        "Cl_ce" : -1,
    }
}
for metabolite, reactions in importdict.items():
    if metabolite != "STARCH":
        tempimportmetabolite = addmetabolite(model, metabolite + "_importfvapseudo", compartment = "", multi=False)
        for reaction, stoich in reactions.items():
            model.reactions.get_by_id(reaction+ "_gc_2").add_metabolites({
                tempimportmetabolite: stoich
            })
        tempimportreaction = addreaction(model, metabolite + "_importfvaconstraint", multi=False)
        tempimportreaction.add_metabolites({
            tempimportmetabolite: -1,
        })
        tempimportreaction.lower_bound = -1000
```

### Add accumulation pseudometabolites

```python
for metabolite in importdict:
    if metabolite == "STARCH":
        templinkermetabolite = addmetabolite(model, metabolite + "_linkerfvapseudo", compartment = "", multi=False)
        model.reactions.get_by_id(metabolite + "_p_gc_Linker_1").add_metabolites({
            templinkermetabolite : 1,
        })
        model.reactions.get_by_id(metabolite + "_p_gc_Linker_2").add_metabolites({
            templinkermetabolite : -1,
        })
        templinkerreaction = addreaction(model, metabolite + "_linkerfvaconstraint", multi=False)
        templinkerreaction.add_metabolites({
            templinkermetabolite : 1,
        })
        templinkerreaction.lower_bound = -1000
    else:
        templinkermetabolite = addmetabolite(model, metabolite + "_linkerfvapseudo", compartment = "", multi=False)
        model.reactions.get_by_id(metabolite + "_c_gc_Linker_1").add_metabolites({
            templinkermetabolite : 1,
        })
        model.reactions.get_by_id(metabolite + "_c_gc_Linker_2").add_metabolites({
            templinkermetabolite : -1,
        })
        model.reactions.get_by_id(metabolite + "_v_gc_Linker_1").add_metabolites({
            templinkermetabolite : 1,
        })
        model.reactions.get_by_id(metabolite + "_v_gc_Linker_2").add_metabolites({
            templinkermetabolite : -1,
        })
        
        templinkerreaction = addreaction(model, metabolite + "_linkerfvaconstraint", multi=False)
        templinkerreaction.add_metabolites({
            templinkermetabolite : 1,
        })
        templinkerreaction.lower_bound = -1000
```

### Add synthesis pseudometabolites

```python
compartmentlist = ["p", "c", "v", "m"]
for metabolite in importdict:
    tempsynthesismetabolite = addmetabolite(model, metabolite + "_synthesisfvapseudo", compartment = "", multi=False)
    for compartment in compartmentlist:
        try:
            for reaction in model.metabolites.get_by_id(metabolite + "_" + compartment + "_gc_2").reactions:
                if "Linker" not in reaction.id and "_" + compartment + "_" in reaction.id:
                    reaction.add_metabolites({
                        tempsynthesismetabolite: reaction.get_coefficient(model.metabolites.get_by_id(metabolite + "_" + compartment + "_gc_2")),
                    })
        except:
            pass
    tempsynthesisreaction = addreaction(model, metabolite + "_synthesisfvaconstraint", multi=False)
    tempsynthesisreaction.add_metabolites({
        tempsynthesismetabolite: -1,
    })
    tempsynthesisreaction.lower_bound = -1000
```

- For aperture size, the constraint value will be taken based on an aperture between 15 and 25 therefore the 15+x*10, giving osmolarity of between 1.12M and 5.56M
